```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = {'Company': ['Allied Signal', 'Bankers Trust', 'General Mills', 'ITT Industries', 'J.P.Morgan & Co.', 'Lehman Brothers', 'Marriott', 'MCI', 'Merrill Lynch', 'Microsoft', 'Morgan Stanley', 'Sun Microsystems', 'Travelers', 'US Airways', 'Warner-Lambert'], 
        'Measure X': [24.23, 25.53, 25.41, 24.14, 29.62, 28.25, 25.81, 24.39, 40.26, 32.95, 91.36, 25.99, 39.42, 26.71, 35]}

df = pd.DataFrame(data)
print(df)

plt.boxplot(df['Measure X'])
plt.title('Finding Outliners')
plt.show()
print('As you can see the form the above boxplot diagram 91.36 is the outlier')
print('We successfully fond the outliners using Boxplot')

# Calculating the mean 
mean = df['Measure X'].mean()
#Calculating the standard deviation 
std = df["Measure X"].std()
#Calculating the variance 
var = df["Measure X"].var()

#Printingall the values 
print('Mean:',mean)
print('Standard Deviation:',std)
print('Variance:',var)
```
